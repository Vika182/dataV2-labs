```python
#we see that only GrLivArea (Above grade (ground) living area in square feet) correlates with the price significantly
df['SalePrice'].corr(df['GrLivArea'])
#we can say that the more the area the more the price (unlike the Lot or Garage area). 
#Though Garage can play a second role
```

```python
df_mean = df[["SalePrice", "Neighborhood"]].groupby(["Neighborhood"]).agg("mean")
df_mean.plot.barh()
#three neighborhoods have significantly more expensive homes than the rest
# they are: NoRidge, NridgHt and StoneBr (more than 300 thousand)
```
